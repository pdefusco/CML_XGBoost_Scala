```scala
import org.apache.spark.sql._
import org.apache.spark.sql.types

val spark = {
  NotebookSparkSession.builder()
    .master("local[*]")
    .getOrCreate()
}
```

```scala
val custintDF = spark.read.option("header","true").csv("customer_interactions.csv")
```

```scala
custintDF.schema
```

```scala
custintDF.show
```

```scala
import spark.implicits._
val intDF = custintDF.withColumn("recency", $"recency".cast(types.IntegerType))
.withColumn("history", $"history".cast(types.FloatType))
.withColumn("used_discount", $"used_discount".cast(types.IntegerType))
.withColumn("used_bogo", $"used_bogo".cast(types.IntegerType))
.withColumn("is_referral", $"is_referral".cast(types.IntegerType))
.withColumn("conversion", $"conversion".cast(types.IntegerType))
.withColumn("score", $"score".cast(types.FloatType))
```

```scala
intDF.schema
```

```scala
import vegas._
import vegas.render.WindowRenderer._
import vegas.sparkExt._
```

```scala
val plotDF = intDF.groupBy("offer").count
```

```scala
plotDF.schema
```

```scala
//implicit val render = vegas.render.ShowHTML(publish(_))
```

```scala
Vegas("Offer Count")
    .withDataFrame(plotDF)
    .encodeX("offer", Ordinal)
    .encodeY("count", Quantitative)
    .mark(Bar)
    .show
```
